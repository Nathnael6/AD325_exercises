```python
class MyHashMap:

    def __init__(self):
        # B is going to be the number of buckets and we are going to use prime 
        # to reduce the clustering, and each inner list holds key, value pairs.
        self._B = 10007
        self._buckets = [[] for _ in range(self._B)]

    def _idx(self, key: int) -> int:
        """ 
        this is going to hash the key to a bucket index. 
        and for integers, modulo is simple and effective 
        """
        return key % self._B

    def put(self, key: int, value: int) -> None:
        """ 
        so for the Put, it is going to insert a key, value pair 
        and if the key already exists, it is going to update the value
        """
        bucket = self._buckets[self._idx(key)]

        # I am now going to check if key exists then update the value 
        for pair in bucket:
            if pair[0] == key:
                pair[1] = value
                return

        # otherwise append a new key,value
        bucket.append([key, value])

    def get(self, key: int) -> int:
        """  
        this is going to return the value for key or -1 if not found 
        """
        bucket = self._buckets[self._idx(key)]

        for k, v in bucket:
            if k == key:
                return v

        return -1

    def remove(self, key: int) -> None:
        """
        this is going to remove the key,value pair if they exist 
        """
        bucket = self._buckets[self._idx(key)]

        for i, (k, _) in enumerate(bucket):
            if k == key:
                bucket.pop(i)
                return
```
